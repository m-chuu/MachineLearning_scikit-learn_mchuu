```python
# Load the Excel sheet, skipping the first 4 rows (so row 5 becomes the first)
churn = pd.read_csv("education_career_success.csv")


churn
```

```python
churn
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
x = churn.drop(['Starting_Salary','Student_ID','Years_to_Promotion','Work_Life_Balance','Career_Satisfaction','Job_Offers','Networking_Score','Soft_Skills_Score','Certifications','Internships_Completed','Projects_Completed','High_School_GPA', 'SAT_Score','Gender','Field_of_Study','Current_Job_Level','Entrepreneurship'], axis="columns") # dropping the output column to create the inputs (features)
y = churn["Starting_Salary"]
```

```python
model.fit(x,y) # this trains the model
model.score(x,y) # this scores the model
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
xtest
```

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(xtrain, ytrain) # using only the training data
model.score(xtest, ytest) # using unseen testing data to score
```

```python
hire = pd.DataFrame([[29, 181, 3.86]], columns=['Age', 'University_Ranking', 'University_GPA'])
model.predict(hire)
print(model.predict(hire))
```

```python
print(model.predict_proba(hire))
```

```python
from sklearn.inspection import permutation_importance

permutation_score = permutation_importance(model, xtrain, ytrain, n_repeats=10)

np.vstack((x.columns, permutation_score.importances_mean)).T
```
